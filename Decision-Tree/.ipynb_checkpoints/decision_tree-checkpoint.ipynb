```python
# importing libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
import graphviz
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
# loading phishing data set
df = pd.read_csv('phishing.csv', sep=';')
df
```

```python
df.describe()
```

```python
# Preprocessing the data

#save column into a list
colnames = df.columns
colnames
```

`Note`: As the interpretation of the -1’s and 1’s in the Result column seems to be missing from the document, it may be helpful to know that a ‘1‘ corresponds to a phishing site and a ‘-1’ to a legitimate site.

```python
# Split the dataset into features and target
X = df.drop(columns=['Result'])
y = df['Result'].map({1: 'Phishing', -1: 'Legitimate'}) # Map values to proper labels
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Train a decision tree classifier
classifier = tree.DecisionTreeClassifier(max_depth=3)
classifier.fit(X_train, y_train)
```

### 1. An image of the final decision tree

```python
# Visualize the decision tree using Graphviz
dot_data = tree.export_graphviz(
    classifier,
    out_file=None,
    feature_names=X.columns,
    class_names=['Legitimate','Phishing'],
    filled=True, 
    rounded=True,
    special_characters=True)

graph = graphviz.Source(dot_data)

graph
```

```python
#rules of the decision tree in text format
r = tree.export_text(classifier, feature_names=list(X_train.columns))
print(r)
```

### 2. Written instructions for an internet analyst to make the decision of whether the website is likely to be a phishing site or not. The instructions must match one-to-one with your decision tree

Here are the written instructions that match one-to-one with the decision tree in the graph:

Step 1. Root Node / SSLfinal_State Check:
* **Condition:** Check if the `SSLfinal_State` is less than or equal to 0.5.
* If True, proceed to Left Child Node.
* If False, proceed to Right Child Node.

Step 2. Left Child Node / URL_of_Anchor Check:
* **Condition:** Check if the `URL_of_Anchor` is less than or equal to -0.5.
* If True, classify the website as Legitimate.
* If False, proceed to Left-Right Child Node.

Step 3. Left-Right Child Node / Prefix_Suffix Check:
* **Condition:** Check if the `Prefix_Suffix` is less than or equal to 0.
* If True, proceed to Left-Right-Left Child Node.
* If False, proceed to Left-Right-Right Child Node.

Step 4. Right-Child Node / URL_of_Anchor Check:
* **Condition:** Check if the `URL_of_Anchor` is less than or equal to -0.5.
* If True, proceed to Right-Left Child Node.
* If False, proceed to Right-Right Child Node.

Step 5. Right-Left Child Node / Prefix_Suffix Check:
* **Condition:** Check if the `Prefix_Suffix` is less than or equal to 0.
* If True, proceed to Right-Left-Left Child Node.
* If False, class the website as Phishing.

Step 6. Right-Right Child Node / web_traffic Check:
* **Condition:** Check if the `web_traffic` is less than or equal to 0.5.
* If True, proceed to Right-Right-Left Child Node.
* If False, proceed to Right-Right-Right Child Node.

### 3. The accuracy estimate (percentage of correct classification) of decision tree

```python
# predict
y_pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy calculated: {accuracy:.4f}")
```

```python
# print classification report
print(classification_report(y_test, y_pred))
```

```python
# output confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(f"confusion matrix:\n{cm}")
```

```python
# Cross-validate
# number of folds
k = 10
scores = cross_val_score(estimator=classifier,
X=X_test,
y=y_test,
scoring="accuracy",
cv=k)
print("Accuracies from %d individual folds:" % k)
print(scores)
print("Accuracy calculated using %d-fold cross validation = %.3f" % (k, scores.mean()))
```
